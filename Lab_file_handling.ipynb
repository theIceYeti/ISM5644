#### 2. Assume that there exists a text file called "numbers2.txt" in your local Python folder. This file contains an unknown number of lines, each consisting of a single positive integer. Write a program that reads through the file, finds the largest number in the file and writes this number to a new text file called "largestNumber.txt". Test your program by creating a sample "numbers2.txt" file.

```python
largest = 0
i = open('numbers2.txt', 'r')
for line in i:
    num = int(line)
    if num > largest:
        largest = num
largestNumber = open('largestNumber.txt', 'w')
largestNumber.write(str(largest))
print('The largest number is', largest)
```

#### Then, write a program that opens this file, reads the lines and transforms the words with the first letter being capitalized, and the rest being lowercase. The program should write the transformed words into a <u>new text document</u> line by line. Fixing grammar mistakes is optional.
Hint: At the beginning of your program, you can define a function that performs the letter transformations for any given string (i.e., word). Then, you can invoke this function iteratively as you go over each line in the file.

```python
def transform(word):
    return word[0].upper() + word[1:].lower()
x = open('meme.txt','r')
lines = x.readlines()
for i in [open('meme_transformed.txt', 'w')]:
    for line in lines:
        word = line.strip()
        i.write(transform(word) + '\n')
```
