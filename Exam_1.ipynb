#### Q3. Use the cell below to enter your code for this question. Write a program that prompts the user for two words that are of arbitrary lengths. The words can be upper case, lower case, or mixed. Print each word in a separate line with the following conversions:

The longer word should be all capitalized. 
Only the 1st letter of the shorter word should be capitalized.
Note: If the words are equal lengths, pick the 1st word as the longer one.

```python
x = str(input('Enter two words'))
x = x.split(' ')
wordOne = (x[0])
wordTwo = (x[1])
if len(wordOne) >= len(wordTwo):
    wordOne = wordOne.upper()
    wordTwo = wordTwo.capitalize()
    print(wordOne)
    print(wordTwo)
else: 
    wordTwo = wordTwo.upper()
    wordOne = wordOne.capitalize()
    print(wordTwo)
    print(wordOne)
```
