#### 1. To find the average of a sequence in a list, we can sum the items and divide the sum to the number of items in the list. Find the average value in the following list: [9,41,12,3,74,11] and print it out.
#### Note: You should only use “+” and “/” arithmetic operators and a *for* loop – but no other function or method from any packages.

```python
totalList = [9,41,12,3,74,11]
counter = 0
sum = 0
for i in totalList:
    sum += i
    counter += 1
print('The average is:', sum/counter)
```

#### 5. Write a program to calculate and print the sum of all prime numbers from 1 to n (n inclusive) where n is a number input by the user.
#### Hint: Consider using a nested loop. The outer loop can be useful for iterating over the numbers between 1 and n and the inner loop can be used to check if the current number in the iteration is prime or not.

```python
n = int(input('Enter a number:'))
is_Prime = True
numSum = 0
for i in range (2,n+1):
    is_Prime = True
    for j in range (2,i):
        if i % j == 0:
            is_Prime = False
            break
    if is_Prime:
        numSum += i
    
print('The sum of all the prime numbers is:', numSum)
```

#### 12. Assume that there exists a list called my_list. Write a program to check if there are duplicate elements (i.e., if any element appears more than once) in my_list. Test your code with <u>different</u> list inputs and print out the finding (i.e., whether duplicates exist or not).

```python
my_list = [1,3,4,5,7]
duplicate = False
count = 0
for i in my_list:
    if i == i:
        count += 1
    if count > 1:
        duplicate = True
print(duplicate)
```
