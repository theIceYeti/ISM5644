#### 8. Write a program to find and print a five-digit number such that if we put the digit '1' at the beginning (i.e. most left), the resulting six-digit number is 3 times smaller than if we put the digit '1' at the end of the number. Use a break to stop the while loop once a number has been found.

```python
num = 10000

while num <= 99999:
    begining = int("1" + str(num))
    end = int(str(num) + "1")
    if end == 3 * begining:
        print('The five digit number is', num)
        break
    num += 1
```

#### 9.The Collatz conjecture is a conjecture in mathematics that concerns a sequence of positive numbers, ending at 1 using an algorithm. The algorithm is as follows:
* Ask the user for a numeric input
* If the number is odd, multiply by 3 and add 1
* If the number is even, divide by 2
* When the number reaches 1, quit
* At each iteration, print out the number on a new line
#### Implement this algorithm using while and if-else controls.

```python
numInput = int(input('Enter a Number: '))
while numInput > 1:
    if numInput % 2 == 0:
        numInput = numInput / 2
        print(numInput)
    else:
        numInput = numInput*3
        numInput += 1
        print(numInput)
    
```
